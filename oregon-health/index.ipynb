```python
import pandas as pd
```

```python
df_descriptive = pd.read_stata('oregonhie_descriptive_vars.dta')
```

```python
df_inperson = pd.read_stata("oregonhie_inperson_vars.dta")
```

```python
df_inperson['dep_dx_post_lottery_inp'].value_counts()
```

```python
df_inperson['dep_dx_pre_lottery_inp'].value_counts()
```

```python
combined_df = pd.merge(df_descriptive, df_inperson, on = 'person_id')
```

```python
dep_cols = ['dep_dx_pre_lottery_inp', 'dep_dx_post_lottery_inp']
bp_cols = ['bp_dar_inp', 'bp_sar_inp']
er_visits = ['ed_num_incl_probe_inp']
borrow_money = ['borrow_inp']
treatment = ['treatment']

cols = treatment + dep_cols + bp_cols + er_visits + borrow_money
selected_df = combined_df[cols]

rename_cols = ['treatment', 'depression_pre', 'depression_post', 'bp_dias', 'bp_syst', 'er_visits', 'borrow_for_med']
selected_df.columns = rename_cols
```

```python
selected_df.T
```

```python
selected_df.to_csv('./oregon_health_insurance.csv', index = True)
```
